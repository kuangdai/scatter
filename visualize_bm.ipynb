```python
import matplotlib.pyplot as plt
import numpy as np

from background_model import PolynomialModel
```

```python
prem_iso = PolynomialModel("./background_models/prem_iso.bm")
prem_iso_NC = PolynomialModel("./background_models/prem_iso_no_crust.bm")
prem_iso_ND = PolynomialModel("./background_models/prem_iso_no_discontinuity.bm")
```

```python
radii = np.linspace(6371., 0., 20000)
vp = [prem_iso.get_properties(r, "VP") for r in radii]
vp_NC = [prem_iso_NC.get_properties(r, "VP") for r in radii]
vp_ND = [prem_iso_ND.get_properties(r, "VP") for r in radii]
```

```python
fig, ax = plt.subplots(1, 2, dpi=200)
ax[0].plot(vp, radii, label="Original")
ax[0].plot(vp_NC, radii, label="No Crust")
ax[0].plot(vp_ND, radii, label="No Discontinuity")
ax[0].set_xlabel("VP")
ax[0].set_ylabel("Radius")
ax[1].plot(vp, radii)
ax[1].plot(vp_NC, radii)
ax[1].plot(vp_ND, radii)
ax[1].set_ylim(6371 - 1000, 6371)
ax[1].set_xlabel("VP")
ax[0].legend(loc="lower left")
plt.show()
```
